```python
import re

def to_lower_camel_case(col):
    # Hilangkan spasi di depan/belakang
    col = col.strip()
    # Pecah berdasarkan spasi atau underscore
    parts = re.split(r'[ _]+', col)
    # Part pertama lowercase, sisanya Capitalize
    return parts[0].lower() + ''.join(word.capitalize() for word in parts[1:])

# Terapkan ke seluruh kolom dataframe
datacovid.columns = [to_lower_camel_case(col) for col in datacovid.columns]
```

```python
datacovid.columns
```

```python
datacovid = datacovid.drop(['cityOrRegency'], axis=1)
```

```python
datacovid['province'].unique()
```

```python
dataclust = datacovid.groupby('province')[['totalCases','totalDeaths','totalRecovered']].max().reset_index()
```

# ðŸŸ¥ **Cluster 4 â€” Wilayah dengan Kasus Sangat Tinggi (Hotspot)**

ðŸ“Œ **Interpretasi:**

* Fasilitas kesehatan kewalahan (ditunjukkan oleh tingginya kematian)

Kita bisa menyebut cluster ini sebagai **wilayah kritis / epicenter**.



```python
X = dataclust[['totalCases','totalDeaths','totalRecovered']]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4, random_state = 0, init = 'k-means++', n_init='auto', max_iter = 500)
kmeans.fit(X)
clust_kmeans = kmeans.fit_predict(X)
```

```python
dataclust['cluster'] = clust_kmeans
dataclust
```

```python
centroid_kmeans = dataclust.groupby('cluster', as_index=False)[['totalCases','totalDeaths','totalRecovered']].mean()
centroid_kmeans
```

```python
def f(row):
    if row['cluster'] == 0:
        val = 1
    elif row['cluster'] == 3:
        val = 2
    elif row['cluster'] == 2:
        val = 3
    else:
        val = 4
    return val
```

```python
dataclust['cluster'] = dataclust.apply(f, axis=1)
dataclust
```
